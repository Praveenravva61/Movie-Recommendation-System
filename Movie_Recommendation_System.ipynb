# Removing unwanted columns from dataset


```python
movies.columns
```

```python
col = ["movie_id", 'title', "genres", 'keywords','overview', 'cast','crew'] 
```

```python
movies = movies[col]
movies
```

```python
movies.isnull().sum()
```

```python
movies =movies.dropna()
```

```python
def genre(text):
    Genre = []
    for i in ast.literal_eval(text):
        Genre.append(i['name'])
    return Genre


    
```

```python
movies['genres'] = movies['genres'].apply(genre)
movies
```

```python
movies['keywords'] = movies['keywords'].apply(genre)
movies
```

```python
movies['cast'][1]
```

```python
def cast(text):
    c = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter <3:
            c.append(i['name'])
            counter +=1
    return c
```

```python
movies['cast'] = movies['cast'].apply(cast)
movies
```

```python
movies['crew'][1]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies
```

Here we have a problelm in all columns. exalmple "Science Fiction" is complete on word but due to space in between that i consider as to different words. So we Need to remove spaces in between the words in the columns.


#  Removing Space gaps in columns

```python
def Removing_spaces(text):
    text1 = []
    for i in text:
        text1.append(i.replace(' ', ''))
    return text1
    
    
        
```

```python
movies.columns
```

```python
movies['keywords'] =movies['keywords'].apply(Removing_spaces)
movies['cast'] =movies['cast'].apply(Removing_spaces)
movies['crew'] =movies['crew'].apply(Removing_spaces)
movies
```

```python
movies['overview']= movies['overview'].apply(lambda x: x.split())

movies.head()
```

```python
movies['overview'][1]
```

```python
movies.columns
```

```python
movies['tags'] = movies["genres"]+movies["keywords"]+movies["overview"]+movies['cast']+movies['crew']
movies.head()
```

```python
new_movies = movies[['movie_id', 'title', "tags"]]
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

```

```python

def stem(text):
    y = []
    for word in text:  # Assuming text is a list of words
        y.append(ps.stem(word))
    return " ".join(y)  # Join the stemmed words back into a single string

# Apply the stem function to each element in the 'tags' column
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
new_movies['tags'][0]
```

```python

cv = CountVectorizer(max_features= 5000, stop_words = 'english')
vectors = cv.fit_transform(new_movies['tags']).toarray()
vectors[0]
    
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x: x[1])
```

```python

        
def recomended_movies(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[movie_index])), reverse=True, key=lambda x: x[1])
    
    for i in distances[1:6]:  # Skip the first item since it's the movie itself
        print(new_movies.iloc[i[0]].title)
   
```
